#### Добавим функции для подготовки данных

```python
def prepare_year(df):
    df.loc[df['HouseYear']>2019, 'HouseYear']=2019
    return df
```

```python
def prepare_rooms(df):
    df.loc[df['Rooms']==0,'Rooms']=1
    df.loc[df['Rooms']>6, 'Rooms']=df.loc[df['Rooms'].between(1, 6), 'Rooms'].median()
    return df
```

```python
def prepare_square(df):
    df.loc[df['Square'] < 14, 'Square'] = 14
    df.loc[df['Square']>(df['Rooms']*40+40),'Square']=df['Rooms']*40+40
    df.loc[df['Square']<(df['Rooms']*15),'Square']=df['Rooms']*15
    return df
```

```python
def prepare_lifesquare(df):
    koef=df['LifeSquare'].median() / df['Square'].median()
    df['LifeSquare']=df['LifeSquare'].fillna(df['Square']*koef)
    df.loc[(df['Square']) < (df['LifeSquare']),'LifeSquare']=df['Square']*koef
    df.loc[df['LifeSquare'] > 230, 'LifeSquare'] = df.loc[df['LifeSquare'] > 230, 'Square']
    return df
```

```python
def prepare_floor(df):
    df.loc[df['HouseFloor']<=0, 'HouseFloor'] = df['Floor']
    df.loc[df['Floor'] > df['HouseFloor'], 'Floor'] = df['HouseFloor']
    df.loc[df['HouseFloor'] > 60 , 'HouseFloor'] = df['Floor']    
    return df
```

```python
#уж очень плохие данные
def clean_data(df):
    df=df.drop(['Healthcare_1'], axis=1)
    return df
```

```python
def prepare_kitchen(df):
    df['KitchenRatio'] = np.divide(df['KitchenSquare'],df['Square'])
    df.loc[(df['KitchenSquare'] > 250) | (df['KitchenRatio'] > 0.6), 'KitchenSquare']=\
                                        df['Square']*(df['KitchenRatio'].mean())
    df.loc[(df['KitchenSquare'] < 5), 'KitchenSquare'] = 5
    return df
```

```python
def join_stats1(df,stats1,source_df,mode='train'):
    df=pd.merge(df,stats1,on=['DistrictId','Rooms'],how='left')
    if mode=='test':
        df['mean_price']=df['mean_price'].fillna(source_df['mean_price'].mean())
    return df
```

```python
stats1=data.groupby(['DistrictId','Rooms'],as_index=False)[['Price']].mean().rename(columns={'Price':'mean_price'})
```

```python
data=join_stats1(data,stats1,data)
```

```python
def prepare_data(df):
    df=prepare_rooms(df)
    df=prepare_year(df)
    df=prepare_square(df)
    df=prepare_lifesquare(df)
    df=prepare_floor(df)
    df=clean_data(df)
    df=prepare_kitchen(df)
    df=pd.get_dummies(df)
    return df
```

```python
data=prepare_data(data)
```

### Построение модели

```python
#оставил полезные признаки (пробовал разные комбинации)
features = data[['Rooms', 'Square', 'LifeSquare', 'KitchenSquare', 'Floor',
       'HouseFloor', 'HouseYear', 'Ecology_1', 'Social_1', 'Social_2',
       'Social_3', 'Helthcare_2', 'Shops_1', 'mean_price',
        'Ecology_2_A', 'Ecology_2_B', 'Ecology_3_A', 'Ecology_3_B', 'Shops_2_A']]
target = data['Price']
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor as RF
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#параметры уточнял не с первой итерации
params={
        'min_samples_leaf':[2,3,4],
        'n_estimators':[180,200,230],
        'max_depth':[14,15,16]
        }
regr=GridSearchCV(RF(),param_grid=params, scoring='r2', cv=3)
```

```python
regr.fit(X_train, y_train)
```

```python
regr.best_params_
```

```python
regr.best_score_
```

```python
model=RF(n_estimators=180,max_depth=15,min_samples_leaf=4, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_train)
```

```python
y_predv=model.predict(X_valid)
```
